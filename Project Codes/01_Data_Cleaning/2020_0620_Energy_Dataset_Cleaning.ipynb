# Data Cleaning

**Columns**
- Drop NAN and 0 cols
- Remove total load forecast

```python
#Drop nan cols
df.dropna(axis =1, inplace= True,thresh=1)

#Subset price_day_ahead to write to a CSV file. We'll try to outperform these predictions when modeling
price_forecast = df['price day ahead']

#Drop forecast variables
drop = df.filter(regex='ahead|forecast').columns
df.drop(drop, axis = 1, inplace =True)

#Drop cols with only 1 unique value in column
df = df.loc[:,df.apply(pd.Series.nunique) != 1]

#Check
display(df.head())
```

**Missing Values**

```python
#Capture rows with at least one missing value and assign to missing
missing = df[df.isnull().any(axis=1)]

#View missing and number of rows with missing data
print('Number of missing rows: {}\nTotal % missing from data {}%'.format(len(missing),round(len(missing)/len(df),4) * 100))
display(missing)
```

*Resolve Missing Values*

```python
#Apply forward and back fill
df = df.ffill().bfill()

#Check
df[df.isnull().any(axis=1)]
```

**Clean time data**

```python
#Remove +01:00
df.time.replace('[+].*','', inplace=True,regex=True)

#Rename dt_iso to time and set type
df.rename(columns = {'time':'date_time'}, inplace = True)

#Set date_time as index
df.set_index('date_time', inplace=True)

#Check
display(df.head(3))
```
